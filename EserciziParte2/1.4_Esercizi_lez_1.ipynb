### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
cubi = []
for i in range(6):
    cubi.append(i**3)

print(cubi)
```

```python
cubi = [i**3 for i in range(6)]
print(cubi)
```

```python
dispari = []
for i in range(7):
    if(i%2) == 1:
        dispari.append(i)

print(dispari)
```

```python
dispari = [i for i in range(7) if(i%2 == 1)]
print(dispari)
```

```python
primi = []
for i in range(1,10):
    value = True
    for j in range(2,i):
        if(i%j == 0):
            value = False
    
    if(value):
        primi.append(i)
    
print(primi)
```

```python
def isPrime(n):
    value = True
    for j in range(2,n):
        if(n%j == 0):
            value = False
    return value

primi = [i for i in range(1,10) if(isPrime(i))]
print(primi)
```

### **3.  Operazioni su array**

```python
import numpy as np
a = np.arange(1,7)
print(a)

b = a[2:4]
print(b)

c=a[::-1]
print(c)

divisione = a/c
print(divisione)

a = []
for i in range(1,7):
    a.append(i)
print(a)

b = a[2:4]
print(b)
c = a[::-1]
print(c)

divisione = []
for i in range(6):
    divisione.append(a[i]/c[i])
print(divisione)
```
